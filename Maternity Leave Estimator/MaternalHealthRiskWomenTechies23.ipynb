```python
X = pd.get_dummies(df_final, columns=["Chronic_Hypertension", "Family_History", "Placenta_Abruption", "Preterm_Labor", "Multiple_Gestation"], drop_first=True)
```

```python
y = X[['RiskLevel']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
from xgboost import XGBClassifier
classifier = XGBClassifier(n_estimators = 50)
classifier.fit(X_train, y_train)
```

```python
y_train_pred = classifier.predict(X_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_train, y_train_pred)
print(cm)
print('Train Accuracy: {}%'.format(round(accuracy_score(y_train, y_train_pred), 4)*100))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_test = le.fit_transform(y_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Test Accuracy: {}%'.format(round(accuracy_score(y_test, y_pred), 4)*100))
```
