```python
import plotly.express as px
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/WomenTechies23/Dataset/sexism_data.csv')
```

```python
data.head()
```

```python
fig = px.line(x=[1, 2], y=[1, 2])
```

```python
print(fig)
```

```python
px.histogram(data, x="sexist", width=500, height=400)
```

```python
plot = px.violin(x = data['sexist'], y = data['toxicity'], width=500, height=400)
plot.update_layout(
    xaxis_title="Sexist",
    yaxis_title="Toxicity",
)
plot.show()
```

```python
data_f = data[['toxicity', 'sexist']]
fig = px.imshow(data_f.corr(), width=500, height=400)
fig.show()
```

```python
fig = px.box(data, x="classify", y="toxicity", width=500, height=400)
fig.show()
```

```python
fig = px.box(data, x="sexist", y="toxicity", width=500, height=400)
fig.show()
```

```python
fig = px.pie(data, values='toxicity', names='sexist', width=500, height=400)
fig.show()
```

```python
fig = px.density_heatmap(data, x="toxicity", y="sexist", width=500, height=400)
fig.show()
```
