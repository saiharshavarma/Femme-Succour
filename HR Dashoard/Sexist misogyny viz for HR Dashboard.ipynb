```python
data = pd.read_csv('sexism-Data final.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("darkgrid")
plt.rcParams['figure.figsize']=(14,8)
```

```python
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
import matplotlib.pyplot as plt
```

```python
a=sns.countplot('sexist',data=data,palette='muted')
plt.title('count of the comment made to be sexist')
plt.ylabel('COUNT',fontsize=14)
plt.xlabel('sexist comment',fontsize=16)
a.set_xticklabels(a.get_xticklabels(),rotation=90)
```

## Distplot of toxicity

-> histogram with a density curve or kernel density estimate (KDE) for visualizing the distribution of toxicity which is a continuous variable 

```python
sns.distplot(data['toxicity'],color='g') 
```

## Violin Plot To Find the relation between sexist and toxicity

```python
sns.violinplot(x='sexist',y='toxicity',data=data,hue='sexist',palette='Accent')
```

## Correlation of sexist and misogynstic dataset

```python
sns.heatmap(data.corr(),annot=True)
plt.rcParams['figure.figsize']=(5,5)
```

## Box Plot to see median, quartiles and skewness for classification and toxicity

```python
sns.boxplot('classify','toxicity',data=data,palette='magma')
```

## Box Plot to see median, quartiles and skewness for sexist comment and its toxicity

```python
sns.boxplot('sexist','toxicity',data=data)
```

```python
sns.histplot(data['toxicity'])
plt.axvline(data['toxicity'].mean(),color='red')
```
